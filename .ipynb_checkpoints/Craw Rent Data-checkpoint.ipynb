```python
#L·∫•y nhi·ªÅu data
print("B·∫Øt ƒë·∫ßu get data ... ")
url = 'https://nha.chotot.com/thue-bat-dong-san-khanh-hoa?page='
header_added = False
i = 1
while True:
    print("ƒêang ·ªü: "+url+str(i))
    wd.get(url + str(i))
    #ƒëi t√¨m t·ª´ng trang, n·∫øu ƒë√£ h·∫øt th√¨ s·∫Ω ng·ª´ng!
    if(wd.find_elements(By.CLASS_NAME,"NotFound_content__QKoFj")):
        print("ƒë√£ h·∫øt!")
        break
    else:
        #l·∫•y t·∫•t c·∫£ url c·ªßa 1 page
        chotot_urls = [tag.get_attribute("href") for tag in wd.find_elements(By.CLASS_NAME,"AdItem_adItem__2O28x")]
        for link in chotot_urls:
            wd.get(link)
            #S·ª≠ d·ª•ng try v√† except ƒë·ªÉ ch·ªëng b·ªã l·ªói.
            #khi ta ch·∫°y ki·∫øm element v√† ph√°t hi·ªán kh√¥ng c√≥ th√¨ s·∫Ω l·ªói v√† ng·∫Øt ch∆∞∆°ng tr√¨nh
            print('ƒêang get name,status,deposit,funiture')
            name = wd.find_element(By.CLASS_NAME,'AdDecription_adTitle__2I0VE').text
            status = wd.find_element(By.CSS_SELECTOR,'#__next > div > div.container > div.ct-detail.adview > div > div.col-md-8 > div:nth-child(5) > div:nth-child(1) > div > div.media-body.media-middle > span > span.AdParam_adParamValue__1ayWO').text
            try:
                deposit = wd.find_element(By.CSS_SELECTOR,'#__next > div > div.container > div.ct-detail.adview > div > div.col-md-8 > div:nth-child(5) > div:nth-child(8) > div > div.media-body.media-middle > span > span.AdParam_adParamValue__1ayWO').text
            except:
                deposit = 'Kh√¥ng c·∫ßn c·ªçc'
            try:
                funiture = wd.find_element(By.CSS_SELECTOR,'#__next > div > div.container > div.ct-detail.adview > div > div.col-md-8 > div:nth-child(5) > div:nth-child(7) > div > div.media-body.media-middle > span > span.AdParam_adParamValue__1ayWO').text
            except:
                funiture = 'Nh√† tr·ªëng'
            print('ƒêang get ti·ªÅn thu√™ v√† di·ªán t√≠ch...')
            text_inside = [text.text for text in wd.find_elements(By.CLASS_NAME,'AdDecription_price__O6z15')]
            for text in text_inside:
                #text s·∫Ω c√≥ d·∫°ng x tri·ªáu/th√°ng - x m2
                #n√™n ta s·∫Ω t√°ch 2 c√°i t·ª´ d·∫•u - v√† c√≥ ƒë∆∞·ª£c ti·ªÅn thu√™ v√† di·ªán t√≠ch
                text = text.split('-')
                rent = text[0]
                area = text[1]
            #Get discribe
            print('ƒêang get m√¥ t·∫£...')
            describe = [text.text for text in wd.find_elements(By.CLASS_NAME,'AdDecription_adBody__1c8SG')]

            #get Phone number
            print('ƒêang get SƒêT')
            try:
                phone = wd.find_element(By.XPATH,'//*[@id="__next"]/div/div[3]/div[1]/div/div[6]/div/div[2]/div[2]/div[1]/div')
                phone.click()

                phone = wd.find_element(By.XPATH,'//*[@id="__next"]/div/div[3]/div[1]/div/div[6]/div/div[2]/div[2]/div[1]/div/div')
                phone_num = phone.text
            except:
                pass
            print('ƒêang get h√¨nh ·∫£nh')
            #get images
            
            images = []
            #l·∫•y ·∫£nh v√† b·ªè v√†o []
            for _ in wd.find_elements(By.CSS_SELECTOR,'div.slick-track'):
                images_link =  [image.get_attribute("src") for image in wd.find_elements(By.CSS_SELECTOR,'div>img[src$=".jpg"]')]
                images.append(images_link)
            print('ƒêang get ƒë·ªãa ch·ªâ')
            #Get address and locate
            try:
                #ƒê·ªãa ch·ªâ n√†y h∆°i ·∫£o, ƒë·∫ßu ti√™n ta s·∫Ω click v√†o "place-desc-large" sau ƒë√≥ s·∫Ω c√≥ 1 modal m·ªü ra
                # Sau ƒë√≥ s·∫Ω get d√≤ng ch·ªØ l√† ƒë·ªãa ch·ªâ v√†o t·ªça ƒë·ªô tr√™n ƒë√≥
                # S·∫Ω c√≥ m·ªôt s·ªë tr∆∞·ªùng h·ª£p kh√¥ng c√≥ data trong modal cho n√™n m·ªõi c√≥ d√≤ng except ·ªü d∆∞·ªõi
                # Tuy nhi√™n m·ªôt s·ªë tr∆∞·ªùng h·ª£p c√≥ data nh∆∞ng v·∫´n kh√¥ng get :D, m√¨nh ch·ªãu
                add = wd.find_element(By.CLASS_NAME,'fz13')
                add_outside = add.text
                add.click()
                #click xong sleep 10s ƒë·ªÉ modal m·ªü l√™n
                time.sleep(10)
                wd.switch_to.frame(1)
                add = wd.find_element(By.CLASS_NAME,'place-desc-large').text
                print(add)
            except:
                #C√°c ƒë·ªãa ch·ªâ b√™n ngo√†i s·∫Ω c√≥ th√™m d√≤ng ch·ªØ xem b·∫£n ƒë·ªì n√™n d√πng strip() c·∫Øt n√≥ ƒëi
                add = add_outside.strip('Xem b·∫£n ƒë·ªì')
                print(add)
            data = []
            #append data... ch·∫Øc s·∫Ω c√≥ c√°ch append ng·∫Øn g·ªçn h∆°n :') 
            data.append(link)
            data.append(name)
            data.append(status)
            data.append(deposit)
            data.append(funiture)
            data.append(rent)
            data.append(area)
            data.append(describe)
            data.append(phone_num)
            data.append(images)
            data.append(add)
            #L∆∞u v√†o file
            #N√™n l√†m b·∫±ng jupyter ch·ª© kh√¥ng ph·∫£i colab
            with open("D:/CSV/final.csv", 'a', newline='', encoding="utf-8") as fp:
                f = csv.writer(fp)
                if not header_added:
                    f.writerow(['Link','Name','Status','Deposit','Funiture_status','Rent', 'Area', 'Describe','Phone_number','Images','Address'])
                    header_added = True
                f.writerows([data])
        i+=1
print('Done!')
wd.quit()
```

```python
#Data cleaning. Sau khi get ƒë∆∞·ª£c data v·ªÅ, data s·∫Ω r·∫•t kh√≥ nh√¨n v√† kh√¥ng ·ªïn.
#N√™n ta ph·∫£i clean data cho ƒë·∫πp ƒë·ªÉ ƒë∆∞a v√†o csv cu·ªëi c√πng
pd.set_option('display.max_rows',1000)
data2 = pd.read_csv('D:/CSV/final_3.csv')


#data t·ª´ m√¥ t·∫£ v√† images s·∫Ω c√≥ [] n√™n ta c·∫Øt n√≥ ƒëi
data2['Describe'] = data2['Describe'].astype(str).str.strip('[]')
data2['Images'] = data2['Images'].astype(str).str.strip('[]')

# data2.loc[data2['Funiture_status'].str.contains('000')] = 'Nh√† tr·ªëng'

# c√≥ m·ªôt s·ªë field kh√¥ng c√≥ trong t√¨nh tr·∫°ng n·ªôi th·∫•t d·∫´n ƒë·∫øn ti·ªÅn c·ªçc s·∫Ω nh·∫£y qua √¥ n√†y
# N√™n t·∫•t c·∫£ nh·ªØng field kh√¥ng c√≥ ƒë√≥ ta thay b·∫±ng nh√† tr·ªëng
data2['Funiture_status'] = data2['Funiture_status'].apply(lambda x: 'Nh√† tr·ªëng' if '000' in x else x )

data2['Phone_number'] = '0' + data2['Phone_number'].astype(str)
#apply phone number c√≥ s·ªë 0 ph√≠a tr∆∞·ªõc, c√≥ th·ªÉ kh√¥ng s·ª≠ d·ª•ng
data2['Phone_number'] = data2['Phone_number'].apply('="{}"'.format)
data2 = pd.DataFrame(data2,dtype = object)

new = data2.to_csv('D:/CSV/rent_data_final.csv',index = False)
#D√πng ƒë·ªÉ l√†m data nh·∫π h∆°n. v√≠ d·ª• data c√≥ 100kb => >99kb. N√™n th·ª≠
# data2.info()
# data2['Funiture_status'].value_counts()
# data2['Funiture_status'] = data2['Funiture_status'].map({'Nh√† tr·ªëng':'Yes','N·ªôi th·∫•t ƒë·∫ßy ƒë·ªß': 'No'})

data2




```
